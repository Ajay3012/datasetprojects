```python
import seaborn as sns
sns.get_dataset_names()
```

```python
import matplotlib.pyplot as plt
```

```python
t=sns.load_dataset('titanic')
t.head(25)
```

```python
plt.Figure(figsize=(8,6))
sns.countplot(x='class',data=t)
plt.title('passenger count by time')
```

```python
plt.Figure(figsize=(8,6))
sns.barplot(x='class',y='pclass',data=t)         #TO DISPLAY THE FREQUENCY OF COUNT OF OBSERVATION IN EACH CATEGORICAL BIN , IT SHOWS HOW MANY 
plt.title('passenger count by time')             #ITEM FALLS IN EACH CATEGORY
```

```python
plt.figure(figsize=(7,8))
sns.histplot(t['age'],bins=25,kde=True,color='orange')
plt.title('age distribution of passanger')
plt.show()
```

```python
counts=t['survived'].value_counts()       # ONLY FOR CATEGORICAL
counts
```

```python
survival_counts=t['survived'].value_counts()
plt.Figure(figsize=(8,7))
plt.pie(survival_counts,labels=['did not survived','survived'],colors=['blue','green'],autopct='%1.1f%%')
plt.title('survivsl rate on titanic')
plt.show()
#TO SHOW PROPOSITIONS OR PERCENTAGES OF DIFFERENT CATEGORIAL IN A WHOLE. SMOT-SYNTHETIC MINORITY OVERSAMPLING TECHNIQUES
```

```python
plt.figure(figsize=(7,8))
sns.boxplot(x='fare',data=t,palette='Set1')
plt.title('fare Distribution by passenger class')
plt.xlabel('class')
plt.ylabel('fare')
plt.show()
```

```python
plt.figure(figsize=(6,8))
sns.scatterplot(x='age',y='fare',hue='survived',style='sex',data=t,palette='rainbow')
plt.title('Age vs fare with survival and Gender')
plt.xlabel('age')
plt.ylabel('fare')
plt.legend(title='survived')
plt.show()
```

```python
plt.figure(figsize=(7,8))
sns.boxplot(x='class', y='fare', data=t, palette='Set1')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

```python
numerical=t[['survived','pclass','age','sibsp','parch','fare']].corr()
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(numerical,annot=True,cmap='coolwarm',center=3)
plt.title('corelation Heatmap of titanic dataset')       #0.75       -0.75       0.075
                                                        #  (+)CORR   (-)CORR     NO CORR
```

```python
plt.figure(figsize=(7,8))
sns.violinplot(x='class',y='age',hue='survived',data=t,split=False,palette='Set2')
plt.title('age distribution by class and survival')
plt.xlabel('class')
plt.ylabel('Age')
plt.legend(title='survived')
```

```python
#
#sns.kdeplot(data=t[''],dropna(),color='green',fill='True')
sns.kdeplot(data=t['sibsp'].dropna(),color='blue',fill=True)
plt.title('Age Distribution')
plt.xlabel('fare')
plt.ylabel('density')
```

```python
sns.kdeplot(data=t['parch'].dropna(),color='blue',fill=True)
plt.title('Age Distribution')
plt.xlabel('fare')
plt.ylabel('density')
```

```python
sns.pairplot(t)
```

```python
sns.pairplot(t[['age','fare','class','survived']])
```

```python

plt.figure(figsize=(6,6))
sns.swarmplot(x='class',y='age',hue='survived',data=t, palette='Set2',dodge=True)
plt.title('Age Distributionby class and survival(swarm plot)')
plt.xlabel('class')
plt.ylabel('Age')
plt.legend(title='survived')
plt.show()
```

```python
g = sns.FacetGrid(t,col='class',row='survived',margin_titles=True)
g.map(sns.histplot,'age',bins=20,color='blue')
g.fig.suptitle('Age Distribution by class and survived' ,y=1.03)
```
